```python
test_df = final_data.copy()
test_df.columns = proper_column_names
```

```python
test_df.drop(0, inplace = True)
```

```python
test_df['Start Date'] = pd.to_datetime(test_df['Start Date'], format="%d-%m-%Y %H:%M")
test_df['End Date'] = pd.to_datetime(test_df['End Date'], format="%d-%m-%Y %H:%M")

# Compute response time (in minutes)
test_df['response_time_minutes'] = (test_df['End Date'] - test_df['Start Date']).dt.total_seconds() / 60
```

```python
num_cat = ['Beer', 'Flavored/Mixed Beer','Non-Alcoholic Beers', 'Hard Ciders','Hard Kombucha', 'Wine', 'Hard Beverage', 'Distilled Spirits','Domestic Light Beer', 'Domestic Regular Beer','Low Calorie Carb Beer', 'Low Calorie Carb Beer', 'Domestic Craft', 'Imported Mexican Beer', 'Other Imported Beer', 'Non-Alcoholic Beer']
categorical_features = ['Urban/Rural','Concept_Replacement Product']
```

```python
df = test_df.copy()
for col in num_cat:
    df[col] = pd.to_numeric(df[col],errors='coerce')

df_transformed = pd.get_dummies(df, columns=categorical_features,dtype=float)
df_transformed
```

```python
df_transformed["Q18"] = df_transformed["Q18A"]+ " "+ df_transformed["Q18B"]+" " + df_transformed["Q18C"]
df_transformed.drop(["Q18A","Q18B","Q18C"], axis = 1, inplace  = True)
```

```python
missing_list=[]

for col in df_transformed.columns:
    if df_transformed[col].isna().sum() > 0:
        print(col)
        missing_list.append(col)


df_transformed[missing_list] = df_transformed[missing_list].fillna("no response")
```

```python
import re
import spacy

nlp = spacy.load("en_core_web_sm")

def clean_text(text):
    if pd.isna(text):  # Handle NaN
        return "No Response"
    
    text = text.lower()  # Convert to lowercase
    text = re.sub(r'\s+', ' ', text)  # Remove extra spaces
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)  # Remove special characters
    
    # Lemmatization using spaCy
    doc = nlp(text)
    text = " ".join([token.lemma_ for token in doc if not token.is_stop])  # Remove stopwords and lemmatize
    
    return text
```

```python
text_col = missing_list
for col in text_col:
    df_transformed[col] = df_transformed[col].apply(clean_text)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

vectorizers = {}
bow_features = []

# Ensure text columns do not have missing values

for col in text_col:
    vectorizer = CountVectorizer(max_features=3)  # Adjust max_features as needed
    bow_matrix = vectorizer.fit_transform(df_transformed[col])

    vectorizers[col] = vectorizer  # Store vectorizer for later use (optional)

    # Convert BoW matrix to DataFrame
    bow_df = pd.DataFrame(bow_matrix.toarray(), 
                          columns=[f"{col}_{word}" for word in vectorizer.get_feature_names_out()])

    # Reset index to avoid merging issues
    bow_df.reset_index(drop=True, inplace=True)
    
    bow_features.append(bow_df)

# Merge all BoW features into a single DataFrame
bow_df_final = pd.concat(bow_features, axis=1)

# Reset index before merging to avoid shape mismatches
df_transformed.reset_index(drop=True, inplace=True)
bow_df_final.reset_index(drop=True, inplace=True)

# Combine with original DataFrame (excluding text columns)
df_bow = pd.concat([df_transformed.drop(columns=text_col), bow_df_final], axis=1)

# Display final DataFrame shape
print(df_bow.shape)
```
